```python
#9) Encourage Patrons to check out books - generate a report of showing 10 patrons who have - checked out the fewest books. 
cursor.execute("SELECT p.PatronID, p.firstName, COUNT(l.LoanID) AS BooksCheckedOut FROM Patrons as p LEFT JOIN Loans as l ON p.PatronID = l.PatronID GROUP BY p.PatronID, p.firstName ORDER BY BooksCheckedOut ASC LIMIT 10")
result = cursor.fetchall()
print("Top 10 patrons with the fewest books checked out:")
for row in result:
    print(f"PatronID: {row[0]}, Firstname: {row[1]}, Books Checked Out: {row[2]}")
```

```python
#-- 10) Find books to feature for an event -- create a list of books from 1890s -- that are currently available 
cursor.execute("SELECT p.PatronID, p.firstName, COUNT(l.LoanID) AS BooksCheckedOut FROM Patrons as p LEFT JOIN Loans as l ON p.PatronID = l.PatronID GROUP BY p.PatronID, p.firstName ORDER BY BooksCheckedOut ASC LIMIT 10")
result = cursor.fetchall()
print("Top 10 patrons with the fewest books checked out:")
for row in result:
    print(f"PatronID: {row[0]}, Firstname: {row[1]}, Books Checked Out: {row[2]}")
```
